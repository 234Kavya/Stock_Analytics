**Data Collection**
The first step in doing so is to create a data set that has all the information that is potentially
useful in predicting stock returns.

````python
#for data manipulation
import numpy as np
import pandas as pd
````

````python
#for web scrapping
from bs4 import BeautifulSoup
import requests
````

````python
#for downloading Stock Data using Yahoo Finance
!pip install yfinance
import yfinance as yf
````

````python
#Scrape Nifty 50 companies from Wikipedia
url = 'https://en.wikipedia.org/wiki/NIFTY_50'
#Accessing the Wikipedia Page
response = requests.get(url)
#Parsing HTML Content
soup = BeautifulSoup(response.text , 'html.parser')
table = soup.find('table', {'id' :'constituents'})
````

````python
table
````

````python
#Extracting Data
company_names =[]
symbol = []
sector = []
for row in table.find_all('tr')[1:]:
  head = row.find('th')
  company_name = head.text.strip()
  company_names.append(company_name)
  cols = row.find_all('td')
  sym =  cols[0].text.strip()+ '.NS'
  symbol.append(sym)
  sec = cols[1].text.strip()
  sector.append(sec)
````

````python
company_names
````

````python
sector
````

````python
symbol
````

````python
#Data Retrieval
data = {}
invalid_tickers = []
for ticker in symbol:
  stock_data = yf.download(ticker , start = '2015-01-01' , end = '2023-12-31')
  if not stock_data.empty:
    stock_data['Ticker'] = ticker
    data[ticker] = stock_data
  else:
    invalid_tickers.append(ticker)
````

````python
#Merging data from all tickers into a single DataFrame.
if data:
    all_data = pd.concat(data.values())
else:
    all_data = pd.DataFrame()
````

````python
all_data
````

````python
# Create a DataFrame
if not all_data.empty:
  all_data = all_data.reset_index()
  all_data['Date'] = pd.to_datetime(all_data['Date'])
````

````python
all_data = all_data[(all_data['Date'].dt.year >= 2015) & (all_data['Date'].dt.year <= 2023)]
all_data.to_csv('nifty_50_stock_prices_2015_2023.csv', index=False)
````
